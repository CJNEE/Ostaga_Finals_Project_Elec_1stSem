```python
import pandas as pd
import warnings
import matplotlib.pyplot as plt
import numpy as np
warnings.filterwarnings('ignore')

fam = pd.read_csv("Family Income and Expenditure.csv")
y = fam['Total Household Income']
```

```python
fam
```

```python
fam.columns
```

```python
y
```

```python
features = ['Total Rice Expenditure', 'Meat Expenditure', 'Total Fish and  marine products Expenditure', 'Fruit Expenditure', 'Vegetables Expenditure', 'Tobacco Expenditure',  'Alcoholic Beverages Expenditure','Clothing, Footwear and Other Wear Expenditure','Housing and water Expenditure', 'Imputed House Rental Value',
       'Medical Care Expenditure', 'Transportation Expenditure','Education Expenditure']

X = fam[features]
```

```python
X
```

```python
fam
```

<h1>Decisition Tree</h1>

```python
from sklearn.tree import DecisionTreeRegressor

my_model = DecisionTreeRegressor()
my_model.fit(X, y)
```

```python
predictions = my_model.predict(X)
```

```python
predictions
```

<h1>TESTING THE ACCURACY OF PREDICTIONS</h1>

```python
from sklearn.metrics import r2_score

r2 = r2_score(y, predictions)
print("R² Score:", r2)
print(f"The model variation is at {r2:.1%}") 
```

<h1>Prediction Using Linear Regression</h1>

```python
from sklearn.linear_model import LinearRegression
```

```python
linear_model = LinearRegression()
linear_model.fit(X,y)
```

```python
linear_model_predisct = linear_model.predict(X)
```

```python
r2 = r2_score(y, linear_model_predisct)
print("R² Score:", r2)
print(f"The model variation is at {r2:.1%}") 
```

```python
fam_corr = fam['Total Household Income'].corr(fam['Total Food Expenditure'])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
corr_columns = [
    'Total Household Income', 'Total Food Expenditure'
]
corr_matrix = fam[corr_columns].corr()
plt.figure(figsize=(12,10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap of Total Household Income and Total Food Expenditure')
plt.show()
```
